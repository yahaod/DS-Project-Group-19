```python
from pytrends.request import TrendReq

pytrends = TrendReq(hl='en-US')

# kw_list = ["Ixom", "/g/1tgv3y0d", "Cleveland Bay Chemicals", "/m/0gpgjw", 
#            "/g/11b7cbc2fp", "/g/11f1kzvhhp", "/m/0n5vjp4"]
alum_list = ["Ixom", "/g/1tgv3y0d", "Cleveland Bay Chemicals", "/m/0gpgjw", "/g/11b7cbc2fp"]
pac23_list = ["Ixom","/m/0gpgjw","/g/11b7cbc2fp"]
pac10_list = ["Ixom","/g/11b7cbc2fp"]
sodhyp_list = ["Ixom","/g/1tgv3y0d","/g/11b7cbc2fp"]
mhl_list = ["Ixom","/g/11f1kzvhhp","/m/0n5vjp4"]
pytrends.build_payload(alum_list, cat=12, timeframe='2015-03-01 2022-03-01', geo='AU', gprop='')
alum_data_after_2015 = pytrends.interest_over_time()
alum_data_after_2015.rename(columns = {'/g/1tgv3y0d':'Coogee Chemicals','/m/0gpgjw':'Nalco Water', 
                                       '/g/11b7cbc2fp':'Omega Chemicals'}, inplace = True)
# , '/g/11f1kzvhhp':'Grenof', '/m/0n5vjp4':'Calix'
pytrends.build_payload(pac23_list, cat=12, timeframe='2015-03-01 2022-03-01', geo='AU', gprop='')
pac23_data_after_2015 = pytrends.interest_over_time()
pac23_data_after_2015.rename(columns = {'/m/0gpgjw':'Nalco Water', '/g/11b7cbc2fp':'Omega Chemicals'}, inplace = True)

pytrends.build_payload(pac10_list, cat=12, timeframe='2015-03-01 2022-03-01', geo='AU', gprop='')
pac10_data_after_2015 = pytrends.interest_over_time()
pac10_data_after_2015.rename(columns = {'/g/11b7cbc2fp':'Omega Chemicals'}, inplace = True)

pytrends.build_payload(sodhyp_list, cat=12, timeframe='2015-03-01 2022-03-01', geo='AU', gprop='')
sodhyp_data_after_2015 = pytrends.interest_over_time()
sodhyp_data_after_2015.rename(columns = {'/g/11b7cbc2fp':'Omega Chemicals', 
                                         '/g/1tgv3y0d':'Coogee Chemicals'}, inplace = True)

pytrends.build_payload(mhl_list, cat=12, timeframe='2015-03-01 2022-03-01', geo='AU', gprop='')
mhl_data_after_2015 = pytrends.interest_over_time()
mhl_data_after_2015.rename(columns = {'/g/11f1kzvhhp':'Grenof', '/m/0n5vjp4':'Calix'}, inplace = True)
```

```python
import pandas as pd
alum_after2015 = pd.read_csv('Google Trend/alum trends data after 2015.csv')
pac23_after2015 = pd.read_csv('Google Trend/pac23 trends data after 2015.csv')
pac10_after2015 = pd.read_csv('Google Trend/pac10 trends data after 2015.csv')
sodhyp_after2015 = pd.read_csv('Google Trend/sodhyp trends data after 2015.csv')
mhl_after2015 = pd.read_csv('Google Trend/mhl trends data after 2015.csv')
alum_after2015['sos'] = 0
pac23_after2015['sos'] = 0
pac10_after2015['sos'] = 0
sodhyp_after2015['sos'] = 0
mhl_after2015['sos'] = 0
```

```python
pac10_after2015
```

```python
def alum_cal_share_search (db):
    db.loc[:,'sos'] = (db['Ixom'])/(db['Ixom']+
                                                             db['Coogee Chemicals']+
                                                             db['Cleveland Bay Chemicals']+
                                                             db['Nalco Water']+
                                                             db['Omega Chemicals'])
    db['sos'] = db['sos'].fillna(0)
    return db

def pac_23_cal_share_search (db):
    db.loc[:,'sos'] = (db['Ixom'])/(db['Ixom']+
                                                             db['Nalco Water']+
                                                             db['Omega Chemicals'])
    db['sos'] = db['sos'].fillna(0)
    return db

def pac_10_cal_share_search (db):
    db.loc[:,'sos'] = (db['Ixom'])/(db['Ixom']+
                                                             db['Omega Chemicals'])
    db['sos'] = db['sos'].fillna(0)
    return db

def SODHYP_cal_share_search (db):
    db.loc[:,'sos'] = (db['Ixom'])/(db['Ixom']+
                                                             db['Coogee Chemicals']+
                                                             db['Omega Chemicals'])
    db['sos'] = db['sos'].fillna(0)
    return db

def MHL_cal_share_search (db):
    db.loc[:,'sos'] = (db['Ixom'])/(db['Ixom']+
                                                             db['Calix']+
                                                             db['Grenof'])
    db['sos'] = db['sos'].fillna(0)
    return db
```

```python
alum_sos = alum_cal_share_search(alum_after2015)
alum_sos.to_csv('share of search/alum_sos_after2015.csv')
pac10_sos = pac_10_cal_share_search(pac10_after2015)
pac10_sos.to_csv('share of search/pac10_sos_after2015.csv')
pac23_sos = pac_23_cal_share_search(pac23_after2015)
pac23_sos.to_csv('share of search/pac23_sos_after2015.csv')
sodhyp_sos = SODHYP_cal_share_search(sodhyp_after2015)
sodhyp_sos.to_csv('share of search/sodhyp_sos_after2015.csv')
mhl_sos = MHL_cal_share_search(mhl_after2015)
mhl_sos.to_csv('share of search/mhl_sos_after2015.csv')
```
