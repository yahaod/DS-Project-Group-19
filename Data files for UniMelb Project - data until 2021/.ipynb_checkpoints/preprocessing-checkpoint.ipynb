```python
import pandas as pd
df = pd.read_excel('Sales Data.xlsx')
df
```

```python
df = df.drop(df[df['Confirmed Quantity']==0].index)
```

```python
df['Sales Unit in L'] = 'undefined'
```

```python
df.head()
```

```python
df.loc[df['Sales Unit']=='DRM', 'Sales Unit in L'] = df['Confirmed Quantity']*200
df.loc[df['Sales Unit']=='BOX', 'Sales Unit in L'] = df['Confirmed Quantity']*1000
df.loc[df['Sales Unit']=='PLY', 'Sales Unit in L'] = df['Confirmed Quantity']*15
df.loc[df['Sales Unit']=='IBC', 'Sales Unit in L'] = df['Confirmed Quantity']*1000
df.loc[df['Sales Unit']=='KL', 'Sales Unit in L'] = df['Confirmed Quantity']*1000
df.loc[df['Sales Unit']=='L', 'Sales Unit in L'] = df['Confirmed Quantity']*1
df.loc[(df['Sales Unit']=='TO') & (df['Material'].str.contains('ALUSUL08')), 'Sales Unit in L'] = df['Confirmed Quantity']*(1000/1.31)
df.loc[(df['Sales Unit']=='TO') & (df['Material'].str.contains('MHL')), 'Sales Unit in L'] = df['Confirmed Quantity']*(1000/1.5)
df.loc[(df['Sales Unit']=='TO') & (df['Material'].str.contains('PAC10')), 'Sales Unit in L'] = df['Confirmed Quantity']*(1000/1.19)
df.loc[(df['Sales Unit']=='TO') & (df['Material'].str.contains('PAC23')), 'Sales Unit in L'] = df['Confirmed Quantity']*(1000/1.34)
df.loc[(df['Sales Unit']=='TO') & (df['Material'].str.contains('SODHYP13')), 'Sales Unit in L'] = df['Confirmed Quantity']*(1000/1.2)
```
