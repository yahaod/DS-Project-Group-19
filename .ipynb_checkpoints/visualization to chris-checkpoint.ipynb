```python
import pandas as pd
import matplotlib.pyplot as plt

sales_data = pd.read_csv('grouped sales data.csv')
sos_after2015 = pd.read_csv('share of search/sos_after2015.csv')
```

```python
sos_after2015['date'] = pd.to_datetime(sos_after2015['date'])
sos_after2015['date'] = sos_after2015['date'].dt.strftime('%Y-%m')
```

```python
pd.set_option('display.float_format', lambda x: '%.3f' % x)
grp_by_time = sales_data.groupby(['Delivery Date'])['Quantity in L'].sum().reset_index()
grp_by_time = grp_by_time.sort_values(by=['Delivery Date'])
```

```python
grp_by_time = grp_by_time.iloc[1: , :]
```

```python
merged_data_after_2015 = pd.merge(grp_by_time, sos_after2015, left_on='Delivery Date', right_on='date', how='inner')
merged_data_after_2015['date'] = pd.to_datetime(merged_data_after_2015['date'])
```

```python
# create figure and axis objects with subplots()
fig,ax = plt.subplots()
# make a plot
ax.plot(merged_data_after_2015['date'], merged_data_after_2015['Quantity in L'], color="red", marker=".")
# set x-axis label
ax.set_xlabel("year",fontsize=14)
# set y-axis label
ax.set_ylabel("sale volume",color="red",fontsize=14)

ax2=ax.twinx()
# make a plot with different y-axis using second axis object
ax2.plot(merged_data_after_2015['date'], merged_data_after_2015['sos_alum'],color="blue",marker=".")
ax2.set_ylabel("share of search",color="blue",fontsize=14)
plt.rcParams["figure.figsize"] = (20,3)
plt.title("Sale VS. Share of Search (All Product)")
plt.show()
```

```python
ALUSUL08_sale_nsw = pd.read_csv('data_state_product/ALUSUL08_sale_nsw.csv')
PAC10LB_sale_nsw = pd.read_csv('data_state_product/PAC10LB_sale_nsw.csv')
PAC23_sale_nsw = pd.read_csv('data_state_product/PAC23_sale_nsw.csv')

sos_nsw = pd.read_csv('share of search/sos_nsw.csv')
```

```python
# create figure and axis objects with subplots()
fig,ax = plt.subplots()
# make a plot
ax.plot(merged_data_alum_nsw['date'], merged_data_alum_nsw['Quantity in L'], color="red", marker=".")
# set x-axis label
ax.set_xlabel("year",fontsize=14)
# set y-axis label
ax.set_ylabel("sale volume",color="red",fontsize=14)

ax2=ax.twinx()
# make a plot with different y-axis using second axis object
ax2.plot(merged_data_alum_nsw['date'], merged_data_alum_nsw['sos_alum'],color="blue",marker=".")
ax2.set_ylabel("share of search",color="blue",fontsize=14)
plt.rcParams["figure.figsize"] = (20,3)
plt.title("Sale VS. Share of Search (Alum in NSW)")
plt.show()
##########################################################################################
# create figure and axis objects with subplots()
fig,ax = plt.subplots()
# make a plot
ax.plot(merged_data_pac10_nsw['date'], merged_data_pac10_nsw['Quantity in L'], color="red", marker=".")
# set x-axis label
ax.set_xlabel("year",fontsize=14)
# set y-axis label
ax.set_ylabel("sale volume",color="red",fontsize=14)

ax2=ax.twinx()
# make a plot with different y-axis using second axis object
ax2.plot(merged_data_pac10_nsw['date'], merged_data_pac10_nsw['sos_pac10'],color="blue",marker=".")
ax2.set_ylabel("share of search",color="blue",fontsize=14)
plt.rcParams["figure.figsize"] = (20,3)
plt.title("Sale VS. Share of Search (pac10 in NSW)")
plt.show()
###########################################################################
# create figure and axis objects with subplots()
fig,ax = plt.subplots()
# make a plot
ax.plot(merged_data_pac23_nsw['date'], merged_data_pac23_nsw['Quantity in L'], color="red", marker=".")
# set x-axis label
ax.set_xlabel("year",fontsize=14)
# set y-axis label
ax.set_ylabel("sale volume",color="red",fontsize=14)

ax2=ax.twinx()
# make a plot with different y-axis using second axis object
ax2.plot(merged_data_pac23_nsw['date'], merged_data_pac23_nsw['sos_pac23'],color="blue",marker=".")
ax2.set_ylabel("share of search",color="blue",fontsize=14)
plt.rcParams["figure.figsize"] = (20,3)
plt.title("Sale VS. Share of Search (pac23 in NSW)")
plt.show()
```
